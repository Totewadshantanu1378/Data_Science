```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
sns.scatterplot(x = "cgpa" , y = "iq" , data = df , hue = "placement")  # the hue helps for linear seperation of placed and not placedplt.
plt.legend(loc = 1) # to laocate the legend onto the right corner of the diagram
```

```python
x = df[['cgpa']]
y = df['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain ,xtest , ytrain , ytest = train_test_split(x , y  ,test_size=0.2 , random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(xtrain , ytrain )
```

```python
lr.score(xtest , ytest )*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy() ,y.to_numpy(), clf=lr) 
# to numpy is used to convert the dataframee into array and clf is the classifierr for logistic (lr object)
```
