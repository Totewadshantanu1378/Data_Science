```python
# we are studying the ordinal encoding
import pandas as pd
```

```python
df = pd.DataFrame({"size" : ['s','m','l','xl','xxl','s','xxl','m']})
df
```

```python
ord_data = [['s','m' ,'l','xl','xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[['size']])
```

```python
df['size_encoded'] = oe.transform(df[['size']])
df
```

```python
ord_data1 = {"s" : 1,"m": 2,"l" : 3,"xl":4 ,"xxl" : 5}   #for mapping the data in increasing order
```

```python
df['size_encoded_mapped'] = df['size'].map(ord_data1)
```
