```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("loan.csv")
df.head()
```

```python
# first of all fill the missing value
df.isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0] , inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Married'].fillna(df['Married'].mode()[0] , inplace=True) # using the mode filling method for categorical data
```

```python
en_data = df[["Married" , "Gender"]]   # this table will have only two columns 
en_data
```

```python
pd.get_dummies(df[["Married" , "Gender"]])   # it will give you the dummy data
# it will get you column splits for categoriacl data 
```

```python
from sklearn.preprocessing import OneHotEncoder   # importing the one hot encoder from sklearn library  
ohe = OneHotEncoder(drop="first")  # drop deletes the first column coming from ohe , four -> two column

# we have to fit the data now

arr = ohe.fit_transform(en_data).toarray()
arr
```

```python
pd.DataFrame(arr, columns=["Married_Yes","Gender_Male"])  # here two columns are there , hence arr should also have two coulmns
```
