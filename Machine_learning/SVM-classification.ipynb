```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x = "cgpa" , y = "iq" ,data = df,hue = "placement")
```

```python
x = df[['cgpa' , 'iq']]
y = df['placement']
```

```python
from sklearn.model_selection import train_test_split
xtrain ,xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel = 'linear')
# if the data is not linearly seperable , you can use the other kernels ilike poly , rbf etc.
sv.fit(xtrain ,ytrain)
```

```python
sv.score(xtest , ytest)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf = sv)
```
