```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as  plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df.drop(columns="EstimatedSalary" ,inplace=True)
```

```python
x = df[['Age']]
y = df['Purchased']
df
```

```python
plt.scatter(x = "Age" , y = "Purchased" ,data = df )
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest ,ytrain , ytest = train_test_split(x,y,test_size=0.2 , random_state=24)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(xtrain , ytrain)
```

```python
lr.score(xtest , ytest)*100
```

```python
plt.scatter(x = "Age" , y = "Purchased" ,data = df )
sns.lineplot(x = "Age" , y = lr.predict(x) ,data = df, color = "red")

# the line will be in s shape morelikely .
```
