```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("placement.csv")
df = df.drop(df.columns[:1] , axis = 1)
df = df.iloc[:,:-1]
df
```

```python
x = df[['cgpa']]
y = df['iq']
```

```python
sns.scatterplot(x = "cgpa",y = "iq" , data = df )
```

```python
from sklearn.model_selection import train_test_split
xtrain ,xtest , ytrain , ytest  = train_test_split(x, y, test_size = 0.2 , random_state = 42)
```

```python
from sklearn.svm import SVR
```

```python
sv = SVR(kernel = "rbf")
sv.fit(xtrain , ytrain )

# the dataset is not suitable thats why showing the minute accuracy iin negative
```

```python
sv.score(xtest , ytest)*100
```

```python
sns.scatterplot(x = "cgpa" ,y = "iq" , data = df)
plt.plot(x, sv.predict(x) , color = "red")

# to inaccurate due to unsuitable dataset 
```
