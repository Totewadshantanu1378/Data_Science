```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("polynomial.csv")
df.head()
```

```python
plt.scatter(df['Temperature (°C)'] , df['Ice Cream Sales (units)'])

# you can see the curve in the scatter plot 
```

```python
x = df[['Temperature (°C)']] 
y =  df['Ice Cream Sales (units)']
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=2) # it is forming the parabola
pf.fit(x )
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain , xtest,ytrain , ytest = train_test_split(x , y , test_size=0.2 , random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(xtrain ,ytrain)
```

```python
lr.score(xtest,ytest)
y_pred = lr.predict(x)
```

```python
plt.scatter(df['Temperature (°C)'] , df['Ice Cream Sales (units)'])
plt.plot(df['Temperature (°C)']  , y_pred , c = "red")
plt.legend(["actual_data","predicted_curve"])
```
