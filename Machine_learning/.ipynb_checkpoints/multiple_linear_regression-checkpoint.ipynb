```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("multiple_linear.csv")
df.head(3)
df = df.drop(columns="Extracurricular Activities")
```

```python
df.isnull().sum()
df
```

```python
# it will print all the features paired graph
sns.pairplot(data = df)
```

```python
sns.heatmap(data = df.corr(), annot=True)
```

```python
x = df.iloc[: , :-1]  #[takes all the rows , leave the last one column andd takes the other columns]
y = df['Performance Index']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train , y_test = train_test_split(x , y , test_size=0.2, random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train , y_train)
```

```python
# accuracy
lr.score(x_test,y_test)
```

```python
lr.coef_
```

```python
y_pred = lr.predict(x_test)
y_pred
```
