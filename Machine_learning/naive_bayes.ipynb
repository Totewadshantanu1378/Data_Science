```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("placement.csv")
df
```

```python
df.isnull().sum()
#no null values present
```

```python
# lets analyze the graph so that we can understand whether it is linearly seperrable or not
plt.figure(figsize=(4,3))
sns.scatterplot(x = "cgpa" , y = "iq",data = df , hue = "placement" )
```

```python
sns.kdeplot(data = df['cgpa'])
```

```python
x = df[['cgpa' , 'iq']]
y = df['placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x , y , test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
```

```python
gnb = GaussianNB()
gnb.fit(xtrain , ytrain)
```

```python
gnb.score(xtest , ytest)*100
```

```python
mnb = MultinomialNB()
mnb.fit(xtrain , ytrain)
```

```python
mnb.score(xtest , ytest)*100
```

```python
bnb = BernoulliNB()
bnb.fit(xtrain , ytrain)
```

```python
bnb.score(xtest , ytest)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy() ,y.to_numpy(), clf=gnb)
```

```python
plot_decision_regions(x.to_numpy() ,y.to_numpy(), clf=mnb)
```

```python
plot_decision_regions(x.to_numpy() ,y.to_numpy(), clf=bnb)  
# here x should contains exacly 2 features so that it can be classified 
# it has 50 - 50 chances so no decision boundary appeared
```
