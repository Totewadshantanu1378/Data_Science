```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# randint generaetes random numbers from 10 to 100
new_data = [np.random.randint(10,100)  for i in range(10000)]
new_data
```

```python
np.mean(new_data)
```

```python
new_table = pd.DataFrame({"new_data" : new_data})
new_table 
```

```python
sns.kdeplot(x = "new_data" , data = new_table)  
```

```python
sam_mean = []
for no_sample in range(50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(new_data))
    sam_mean.append(np.mean(sample_data))
```

```python
np.mean(sam_mean)
```

```python
sample_M = pd.DataFrame({"sample_mean" : sam_mean})
plt.figure(figsize=(4,3))
sns.kdeplot(x = "sample_mean" , data = sample_M)
```

```python
# this code is perforrmed for showing that the mean of sample is approx equals to mean of population 
sns.histplot(x = "sample_mean" , data = sample_M)
```
